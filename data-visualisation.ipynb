# 0 Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

# 1 Importing Data

```python
df = pd.read_csv('output/clean_data.csv')
```

# 1.1 Creating a folder to save the plots

```python
if not os.path.exists('./assets'): os.makedirs('./assets')
```

# 2 Draw histogram(s) of all features

```python
plt.figure(figsize=(20, 20))
for i, col in enumerate(df.columns):
    plt.subplot(4, 4, i+1)
    sns.histplot(df[col], color='blue', kde=True)
plt.savefig('assets/histograms.png')
```

# 3 Draw boxplot(s) (distribution of features)

```python
plt.figure(figsize=(20, 20))
for i, col in enumerate(df.columns):
    plt.subplot(4, 4, i+1)
    sns.boxplot(df[col], color='blue', width=0.5)
plt.savefig('assets/boxplots.png')    
```

# 4 Draw heatmap of correlation matrix (only highly correlated features, i.e. > +/-0.75)

```python
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr()[(df.corr() > 0.75) | (df.corr() < -0.75)], annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.savefig('assets/heatmap.png')
```

# 5 Draw scatterplot(s) with price (target variable)

```python
plt.figure(figsize=(20, 20))
for i, col in enumerate(df.columns):
    plt.subplot(4, 4, i+1)
    sns.scatterplot(x=df[col], y=df['price'], alpha=0.9, s=10, color='blue', marker='x')
plt.savefig('assets/scatterplots.png')    
```

# 6 Draw pairplot(s), - scatterplot of all features

```python
plt.figure(figsize=(20,20))
sns.pairplot(df, corner=True, diag_kind='kde', plot_kws={'alpha': 0.9, 'color': 'blue', 'marker': 'x'})
plt.savefig('assets/pairplots.png')
```
